Com as informações obtidas, tirar uma conclusão sobre os resultados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm


df = pd.read_csv('./Copy of Cópia de japan_heart_attack_dataset.csv')
```

```python
feature = 'Age'
max_age = 40
feature_data = df[feature].dropna()
mu = np.mean(feature_data)
sigma = np.std(feature_data)

x = np.linspace(feature_data.min(), feature_data.max(), 100)
z = (max_age - mu) / sigma

pdf_normal_distribution = norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, pdf_normal_distribution, color='red', label="Distribuição Normal")

plt.title(f'Distribuição de {feature}', size=20)
plt.xlabel(feature, size=15)
plt.ylabel('Densidade de Probabilidade', size=15)
plt.legend()
plt.show()
probabilidade = norm.cdf(z)
print(f"A probabilidade de ter idade menor ou igual a {max_age} anos é: {probabilidade}")
```

```python
cholesterol_data = df['Cholesterol_Level'].dropna()

mu = np.mean(cholesterol_data)
sigma = np.std(cholesterol_data)

x = np.linspace(cholesterol_data.min(), cholesterol_data.max(), 100)
pdf_normal_distribution = norm.pdf(x, loc=mu, scale=sigma)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, pdf_normal_distribution)
ax.set_ylim(0)
ax.set_title('Distribuição Normal do Nível de Colesterol', size=20)
ax.set_ylabel('Densidade de Probabilidade', size=20)
plt.show()
```

```python
bmi_data = df['BMI'].dropna()

mu = np.mean(bmi_data)
sigma = np.std(bmi_data)

x = np.linspace(bmi_data.min(), bmi_data.max(), 100)
pdf_normal_distribution = norm.pdf(x, loc=mu, scale=sigma)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, pdf_normal_distribution)
ax.set_ylim(0)
ax.set_title('Distribuição Normal do IMC', size=20)
ax.set_ylabel('Densidade de Probabilidade', size=20)
plt.show()
```
